## 4. Import the OBJ file

```python
objects = Topology.ByOBJPath(r"C:\IaaC\AIA\AIAStudio\GraphML\hand_boxes5.obj")
print("Objects is a list")
print(objects)
```

```python
cells = []
selectors = []
for object in objects:
    d = Topology.Dictionary(object)
    faces = Topology.Faces(object)
    if len(faces) > 1:
        c = Cell.ByFaces(faces)
        c = Topology.RemoveCollinearEdges(c)
        s = Topology.InternalVertex(c)
        name = Dictionary.ValueAtKey(d, "name")
        if "Joint_1" in name:
            color = "red"
        elif "Joint_2" in name:
            color = "green"
        elif "Joint_3" in name:
            color = "blue"
        else:
            color = "white"
        d = Dictionary.SetValuesAtKeys(d, ["color", "vertex_size"], [color, 20])
        s = Topology.SetDictionary(s, d)
        selectors.append(s)
        cells.append(c)
        print(Dictionary.Keys(d), Dictionary.Values(d))

print(len(cells))
```

```python
hand = Graph.ByTopology(objects[0])
```

```python
vertices = Graph.Vertices(hand)
for v in vertices:
    d = Dictionary.ByKeysValues(["size", "color"], [18, "red"])
    v = Topology.SetDictionary(v, d)
edges = Graph.Edges(hand)
for e in edges:
    d = Dictionary.ByKeysValues(["width", "color"], [4, "black"])
    e = Topology.SetDictionary(e, d)
```
